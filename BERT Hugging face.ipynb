Image Classification

```python
from transformers import pipeline
classifier=pipeline('image-classification')
preds=classifier('/content/amma.jpg')
preds=[{'score':round(pred['score'],4),'label':pred['label']} for pred in preds]
print(*preds,sep="\n")
```

object detection

```python
from transformers import pipeline
classifier=pipeline('object-detection')
preds=classifier('/content/amma.jpg')
preds=[{'score':round(pred['score'],4),'label':pred['label'],'box':pred['box']} for pred in preds]
preds
```

```python
preds=classifier('/content/veena.jpg')
preds=[{'score':round(pred['score'],4),'label':pred['label'],'box':pred['box']} for pred in preds]
preds
```

```python
l=[list(i['box'].values()) for i in preds]
l
```

Image Segmentation

```python
from transformers import pipeline
segmenter=pipeline(task='image-segmentation')
preds=segmenter('/content/amma.jpg')
preds=[{'score':round(pred['score'],4),'label':pred['label'],'box':pred['box']} for pred in preds]
preds
```

```python
from transformers import pipeline
depth_estimator=pipeline(task='depth-estimation')
# Correctly use the depth_estimator pipeline
preds=depth_estimator('/content/veena.jpg')
# Depth estimation output is typically a depth map (PIL Image) and a predicted_depth (tensor)
# We'll print the raw output to inspect it, as it doesn't have 'score', 'label', 'box' keys
print(preds)
```

Text Translation

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained('t5-small')
model = AutoModelForSeq2SeqLM.from_pretrained('t5-small')

text = 'how is the day today'

# Prepare the input text with the T5-specific prefix for translation
input_text = f"translate English to Spanish: {text}"

# Tokenize the input
inputs = tokenizer(input_text, return_tensors='pt')

# Generate the translation
output_tokens = model.generate(
    inputs['input_ids'],
    attention_mask=inputs['attention_mask'],
    max_new_tokens=50 # Limit the length of the generated translation
)

# Decode the generated tokens
output = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
print(output)
```
