BERT Tokenizer

```python
sentence='we are learning bert embeddings'
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
tokenizer
```

```python
tokenizer.vocab.keys()
```

```python
tokenizer.vocab['hello']
```

```python
 tokens=tokenizer.tokenize(sentence)
 tokens
```

```python
tokenizer.vocab['##bed'],tokenizer.vocab['bed']
```

```python
tokens=tokenizer.tokenize('Machinelearning')
tokens
```

```python
 tokens1=['[CLS]']+tokens+['[SEP]']
 tokens1
```

```python
tokens2=['[CLS]']+tokens+['[SEP]']+['[PAD]']+['[PAD]']
tokens2
```

```python
 attention_mask=[1 if i!='[PAD]' else 0 for i in tokens2]
 attention_mask
```

Get Token Ids

```python
 token_ids=tokenizer.convert_tokens_to_ids(tokens2)
 token_ids
```

```python
tokenizer.convert_tokens_to_string(tokens2)
```

```python
tokenizer.convert_ids_to_tokens(token_ids)
```

```python
import torch
ids=torch.tensor(token_ids).unsqueeze(0)
mask=torch.tensor(attention_mask).unsqueeze(0)
ids,mask
```

```python
out=model(ids,attention_mask=mask)
dir(out)
```
