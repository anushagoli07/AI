```python
import os
import gradio as gr
from crewai import Agent,Task,Crew
from crewai_tools import SerperDevTool
```

```python
from google.colab import userdata
os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
os.environ['SERPER_API_KEY'] = userdata.get('SERPER_API_KEY')
```

```python
search_tool=SerperDevTool()
def run_multi_agent(topic):
  teacher=Agent(role='Teacher',
              goal='Explain concepts clearly step by step',
              backstory='A passionate educator dedicated to making complex topics understandable.',
              llm='gpt-4o-mini',
              verbose=True)
  researcher=Agent(role='Researcher',
                 goal='Find lateest information',
                 backstory='A meticulous information gatherer, always seeking the most current and relevant data.',
                 tools=[search_tool],
                 llm='gpt-4o-mini',
                 verbose=True)
  simplifier=Agent(role='Simplifier',
                 goal='Make things easy',
                 backstory='Breaks complex ideas into simple language',
                 llm='gpt-4o-mini',
                 verbose=True)
  student=Agent(role='Student',
              goal='Take notes',
              backstory='writes notes like learner',
              llm='gpt-4o-mini',
              verbose=True)
  examiner=Agent(role='Examiner',
               goal='Grade notes',
               backstory='Tests notes',
               llm='gpt-4o-mini',
               verbose=True)
#topic='what is agents vs agentic AI vs MCP vs generative AI'
  task1=Task(
    description=f'search and find 3 important points about {topic}',
    expected_output='3 clear points',
    agent=researcher
)
  task2=Task(
    description=f'Explain topic in simple steps with examples',
    expected_output='need step by step explaination',
    agent=teacher
)
  task3=Task(
    description=f'simplify the explanation of {topic}',
    expected_output='need to simplify',
    agent=simplifier
)
  task4=Task(
    description=f'write notes about {topic} like a student',
    expected_output='need to write notes',
    agent=student
)
  task5=Task(
    description=f'grade notes of {topic}',
    expected_output='need to grade notes',
    agent=examiner
)
  crew=Crew(
    agents=[teacher,researcher,simplifier,student,examiner],
    tasks=[task1,task2,task3,task4,task5],
    verbose=False
)
  result=crew.kickoff()
  return str(result)
interface=gr.Interface(fn=run_multi_agent,
                       inputs=gr.Textbox(
    label='ENter Topic',
    placeholder='How i AI market in Virginia and DC area right now'
),
                       outputs=gr.Textbox(
    label='Multi-Agent Output'),
                       title='Multi Agent AI Training',
                       description='Enter any job role to see how is the job market right now')
if __name__=='__main__':
  interface.launch()
```
